# Documentation and theory for `BANTS`

---

## Motivating background

\textcolor{red}{1. USE ABSTRACT MARKET PHENOMENA ONLY 2. Apply the nonperturbative renormalisation group like the information bottleneck to look at scaling of optimal effective descriptions 3. Compute the Bayesian evidence for model selection of different possible market structures --- this could even lead to dynamical evidence-weighted ensembles 4. Apply all of the theory developed in this paper in \href{https://github.com/umbralcalc/bants}{\texttt{BANTS}}.}

Consider the hierarchical model

$$
\begin{equation}
{\cal P}(\boldsymbol{\theta}\vert d) = \frac{{\cal L}(d\vert \boldsymbol{\theta})\pi (\boldsymbol{\theta})}{{\cal E}(d)} = \frac{{\cal L}(d\vert \boldsymbol{\theta})}{{\cal E}(d)}\int {\rm d}\boldsymbol{\eta}\,p(\boldsymbol{\theta}\vert \boldsymbol{\eta})\,p(\boldsymbol{\eta}) \,.
\end{equation}
$$
